!!! abstract Objectif
L’objectif du TP est de voir les commandes et syntaxes de base pour faire des recherches simples ou complexes dans une base de données avec le langage SQL.
!!!

## 1.Modélisation par base de données du pokédex 

Il serait possible de créer une base de données référençant tous les pokémons en listant leur nom, leur type, leur forme, leur habitat, leur évolution,… 

Une telle base serait difficile à gérer car il faut pouvoir extraire des informations dans différents langages et le stockage de toutes les traductions des noms, des types, des formes… dans une même table serait quasi impossible. 

La solution est donc de créer plusieurs tables contenant d’une part la liste des entités et une autre leur traduction dans différents langages, ceci pour chaque entité (noms, types, formes…). Des tables intermédiaires peuvent permettre de relier les tables entre elles. 

Voici un extrait des tables et attributs de la base de données que nous étudierons. 


![](https://capytale2.ac-paris.fr/web/sites/default/files/2023/06-26/16-41-46/img_bdd_pokemon.PNG)

Observer les deux tables principales **pokemon_species** (traduire par caractéristiques des pokemons plutôt que par espèce) et **pokemon**.  

!!! question Question 1
Quelles sont les clés primaires pour chaque table ? Comment sont-elles reliées entre elle ? 
!!! 



!!! success Réponse
**pokemon_species** a pour clé primaire *id* et **pokemon** a pour clé primaire *id*, et pour clé étrangère *species_id* associée à la table **pokemon_species**
!!!

Pour réaliser des recherches « complexes », il faut passer par des commandes SQL (doc : http ://sql.sh/cours) .

## 2.Travail sur la base de données veekun-pokedex 

### 1.Sélection sur une table (pokemon_species) 

!!! question Question 2
Déterminer la requête SQL permettant de récupérer les identifiants des pokemon de la generation 1. [151 résultats] 
!!!


```sql
SELECT identifier FROM pokemon_species WHERE generation_id = 1 
```


!!! question Question 3
Déterminer les nomes (identifier) des pokemons de la génération 2 ou 3 qui sont des bébés [10 résultats] 
!!!

```sql
SELECT identifier, generation_id FROM pokemon_species WHERE is_baby=1 and generation_id <=3 and generation_id >=2 
```



!!! question Question 4
Déterminer la requête SQL permettant de récupérer les noms (identifier) des pokemons de la génération 2 en les triant par ordre alphabétique croissant [100 résultats] 
!!!

```sql
SELECT identifier FROM pokemon_species WHERE generation_id=2 ORDER BY identifier ASC 
```



!!! question Question 5
Déterminer la requête SQL permettant de récupérer les noms (identifier) et la génération des pokemons qui sont des bébés. 
!!!


```sql
SELECT identifier,generation_id FROM pokemon_species WHERE is_baby=1 
```


!!! question Question 6
Déterminer la requête SQL permettant de récupérer les noms (identifier) et la génération des pokemons dont le nom (identifier) commence par « k ». 
!!!

```sql
SELECT identifier, generation_id FROM pokemon_species WHERE identifier LIKE "k%" 
```



### 2.Sélection sur plusieurs tables 

!!! question Question 7
Déterminer la requête SQL permettant de récupérer le nom (identifier), l’id et la génération des pokemons (de la table polemon_species) dont la forme (pokemon_shapes) est poisson (fish). [31 résultats] 
!!!


```sql
SELECT identifier,id,generation_id FROM pokemon_species WHERE (shape_id  in (SELECT id FROM pokemon_shapes WHERE identifier LIKE "fish") ) 
```


!!! question Question 8
Déterminer la requête SQL permettant de récupérer les noms (identifier) des pokemons de la génération 1 qui habite dans une forêt (forest). 
!!!


```sql
SELECT identifier FROM pokemon_species WHERE habitat_id =  (SELECT id FROM pokemon_habitats WHERE identifier LIKE "forest") and generation_id = 1 
```


### 3.Opérateurs d’agrégation 

!!! question Question 9
Déterminer la requête SQL permettant de récupérer le nombre de pokemon de chaque génération 
!!!


```sql
SELECT count(identifier) FROM pokemon_species GROUP BY generation_id 
```


!!! question Question 10
Déterminer la requête SQL permettant de connaître le taux minimal, maximal et moyen de capture des pokemons de la génération 1. 
!!!


```sql
SELECT min(capture_rate),max(capture_rate),avg(capture_rate) FROM pokemon_species WHERE generation_id = 1 
```


!!! question Question 11
Déterminer les requêtes SQL permettant de connaître le(s) id et le(s) nom(s) (identifier) des pokemons ayant le taux minimal de capture puis ceux ayant le taux maximal de capture pour la génération 1. 
!!!


```sql
SELECT id, identifier FROM pokemon_species WHERE generation_id = 1 GROUP BY identifier HAVING capture_rate = (SELECT max(capture_rate) FROM  pokemon_species) 
```


!!! question Question 12
Déterminer la requête SQL permettant de récupérer le nombre de pokemons pour chaque habitat. 
!!!

```sql
SELECT count(identifier),habitat_id FROM pokemon_species GROUP BY habitat_id 
```


### 4.Jointure 

L’objectif est d’afficher des résultats en ligne dont les éléments appartiennent à plusieurs tables. 


!!! question Question 13
Déterminer la requête SQL permettant d’afficher les noms en français et taux de capture des pokemons de la génération 1. Le nom de la langue peut être récupéré dans la table languages (identifier=fr) 
!!!


```sql
SELECT name,identifier FROM pokemon_species_names JOIN  pokemon_species ON pokemon_species_id = id  WHERE local_language_id = (SELECT id FROM languages WHERE iso639 = "fr") and generation_id = 1 
```


!!! question Question 14
Déterminer la requête SQL permettant d’afficher les noms en français, le type (type_id) et l’id des pokemons du slot 1. 
!!!

```sql
SELECT name, type_id,pokemon_species_id FROM pokemon_types JOIN pokemon_species_names ON pokemon_species_names.pokemon_species_id = pokemon_types.pokemon_id WHERE pokemon_species_names.local_language_id=5 and pokemon_types.slot=1 
```



!!! question Question 15
Modifier la requête pour afficher les types en français à la place du type_id. 
!!!


```sql
SELECT pokemon_species_names.name, type_names.name FROM pokemon_types JOIN pokemon_species_names ON pokemon_species_names.pokemon_species_id = pokemon_types.pokemon_id  JOIN type_names ON pokemon_types.type_id =type_names.type_id WHERE pokemon_species_names.local_language_id=5 and pokemon_types.slot=1 and type_names.local_language_id=5 
```
