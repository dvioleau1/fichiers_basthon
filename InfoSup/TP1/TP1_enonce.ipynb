# Exercice 8 - Suite de Fibonacci

Leonardo Fibonacci, un mathématicien italien du XIII\ieme\ siècle, a proposé le problème récréatif suivant dans l’un de ses ouvrages.

Un homme met un couple de lapins dans un lieu isolé de tous les côtés par un mur. Combien de couples obtient-on en un an si chaque couple engendre tous les mois un nouveau couple à compter du troisième mois de son existence ?

On en déduit les règles suivantes :
* au mois zéro, il n’y a aucun lapin ;
* au début du premier mois, il y a juste un couple de lapereaux ;
* les lapereaux ne procréent qu'à partir du début du troisième mois ;
* chaque début de mois, tout couple susceptible de procréer engendre effectivement un nouveau couple  de lapereaux ;
* les lapins ne meurent jamais.


Posons $n$ le numéro du mois et $F_n$ le nombre de couples de lapins au mois $n$ :

* $i  = 0$ ; $F_0 = 0$ ; 0 couple de lapins ; 
* $i  = 1$ ; $F_1 = 1$ ; 1 couple de lapins (petits) ; 
* $i  = 2$ ; $F_2 = 1$ ; 1 couple de lapins (moyens)  ; 
* $i  = 3$ ; $F_3 = 2$ ; 2 couples de lapins (1 grand, 1 petit) ; 
* $i  = 4$ ; $F_4 = 3$ ; 3 couples de lapins (1 grand, 1 moyen, 1 petit) ; 
* $i  = 5$ ; $F_5 = 5$ ; 5 couples de lapins (2 grands, 1 moyen, 2 petits) ; 
* $i = n$  ;  $F_n$ couples de lapins                                

!!! question Question 
A partir des itérations de 0 à 5 précédentes, déterminer une relation (dite de récurrence) entre $F_n$, $F_{n-1}$ et $F_{n-2}$.
!!!


!!! question Question
Écrire un programme qui à partir d'un nombre $n$ connu calcule et affiche le nombre de couples de lapins à chaque début de mois, du mois 0 au mois $n$ en utilisant la relation précédente.
!!!


!!! question Question 
Tester le programme pour plusieurs valeurs de $n$ et indiquer au bout de combien de mois aura-t-on plus de 4000 couples de lapins, plus de 40000 couples de lapins ! 
!!!

# Exercice 9 - Remboursement d'emprunt

Lors d'un achat immobilier, il est important de déterminer la façon dont pourra être remboursé un crédit immobilier. En fonction de la somme empruntée, sur combien d'années s'étalera le prêt, quelles seront les mensualités de remboursement du crédit, quel taux pourra-t-on obtenir ?
Il existe de nombreux simulateurs en lignes permettant de faire ces différents calculs. Nous allons ici voir quels sont les codes en place derrière ces simulateurs.

On note $S$ la somme empruntée et $t_a$ le taux annuel (le taux mensuel est calculé simplement par la formule : $t_m=t_a/12$). On note $n$ le nombre de mois pour rembourser la somme empruntée et $m$ la mensualité constante.
Chaque mois, on rembourse une partie du capital emprunté ainsi que des intérêts proportionnel \textbf{au capital restant dû}. C'est là toute la difficulté de ce type d'emprunt. En notant $I_i$ le montant de l'intérêt chaque mois, le capital restant dû le mois $i$ est : $C_i=C_{i-1}-(m-I_i)$ avec $C_0=S$ et $I_i=C_{i-1}.t_m$.

!!! question Question
Ecrire un programme qui affiche le capital restant dû chaque mois et les intérêts du mois pour une somme $S$ fixée, un taux annuel fixé $t_a$ et une mensualité constante $m$. On testera le programme avec $S=400000$, $t_a={0.02}$ et $m={2000}$.
!!!


!!! question Question
Ajouter au programme l'affichage à la fin du nombre de mois nécessaires pour rembourser le prêt (somme empruntée avec intérêts) ainsi que le total des intérêts versés. Améliorer le programme en arrondissant les intérêts au centime d'euro et en gérant la dernière mensualité différemment des autres.
!!!

Dans le cas où le nombre de mois $n$ est fixé, on peut calculer la mensualité par la formule : $$m=\frac{{(1+t_m)}^n}{({1+t_m)}^n-1}.S.t_m$$
Toutefois la mensualité est arrondie au centime d'euro supérieur. La dernière mensualité est donc ainsi réduite.
!!! question Question 
Dans le cas où le nombre de mois $n$ est fixé (la somme empruntée $S$ est également fixée), réaliser un programme qui calcule la mensualité et l'arrondit au centime d'euro supérieur. Calculer la dernière mensualité et indiquer le total des intérêts.
!!!
