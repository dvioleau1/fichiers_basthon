# Exercice 1 - Livret A

On place 1000 euros chaque 1er janvier sur un livret A initialement vide. Le taux de rémunération est de 2\% annuel ce qui signifie que chaque 31 décembre, la somme présente sur le livret au 1er janvier précédent est augmentée de 2\%. 

!!! question Question
Faire un programme qui affiche le montant du livret A à la fin de chacune des 10 premières années. Quel est le gain total ?
!!!


```python
total = 0
for i in range(10):
    total = total + 1000 #ajout de 1000euros au 1er janvier
    total = total*1.02   #ajout des interets au 31 décembre
    print(i+1,total)     #affichage de la somme accumulée

gain_total = total-10000
print(gain_total)
```

# Exercice 4 - Sommes, produits

!!! question Question
Écrire un script qui calcule la somme suivante en prenant deux valeurs quelconques pour $n$ et $p$ entiers naturels : $\sum_{k=n}^{2n}(7k+5)^p$
!!!

```python
n = randint(0,100)
p = randint(0,100)
somme = 0
for k in range(n,2*n+1):
    somme = somme + (7*k+5)**p
print(somme)
```

!!! question Question
Écrire un script qui calcule le produit suivant en prenant $n>0$ et $p \in [1,n]$, entiers naturels : $\prod_{k=1}^{p}\frac{n+1-k}{n}$
!!!

```python
n = randint(1,100)
p = randint(1,n)
produit = 1
for k in range(1,p+1):
    produit = produit * (n+1-k)/n
print(produit)
```

!!! question Question
Ecrire un script qui calcule la valeur de $\sum_{1\leq i,j \leq n} \frac{i^j}{i+j}$  avec $n>1$
!!!

```python
n = randint(1,100)
somme = 0
for i in range(1,n+1):
    for j in range(1,n+1):
        somme = somme + i**j/(i+j)
print(somme)
```

!!!

```python
M = randint(1,1000)
N = 0
somme = 0
while somme < M:
    N = N + 1
    somme = somme + N**2
print(N)
```

# Exercice 5 - Suite

Soit $(u_n)$ la suite définie par $u_0=\frac{1}{2}$ et pour tout $n \in \mathbb{N}$, $u_{n+1}=u_n^2-u_n+1$. On peut montrer que cette suite tend vers $\ell=1$. 

!!! question Question
Ecrire un script qui calcule les termes de la suite $(u_n)$ jusqu'à ce que $u_n$ soit proche de sa limite à $10^{-4}$ près, puis affiche le dernier terme de la suite calculé ainsi que le nombre de termes qu'il a fallu calculer.
!!!

```python
u = 1/2
n = 0
while abs(u-1) > 1e-4:
    u = u**2 -u +1
    n = n + 1
print(u,n)
```

# Exercice 7 - Série de Madhava -- Leibniz
Cette série est définie par $S_n = 4 \sum_{k=0}^n \dfrac{(-1)^k}{2k+1}$.

!!! question Question
Ecrire la suite des instructions permettant de calculer $S_n$. Tester votre programme pour $n=10$, le résultat est $3.232315$.
!!!


```python
Sn = 0
n = 10
for k in range(n+1):
    Sn = Sn + 4*(-1)**k/(2*k+1)
print(Sn)
```

!!! question Question
Tester votre algorithme pour $n=10000$. Vers quelle limite la série semble-t-elle tendre ?
!!!


!!! success Solution
Pour $n=10000$, le résultat est $S_n = 3.1416926435905346$. La suite semble tendre vers $\pi$.
!!!

!!! question Question
Chercher la valeur de $n$ pour avoir deux chiffres significatifs corrects de manière stable (c'est-à-dire que pour tous les $n$ suivants il y ait toujours deux chiffres significatifs).
!!!


```python
n = 0
Sn = 4  #S0, on ne teste pas pour n=0
stable = False
while stable == False:
    n = n + 1
    Sn = Sn + 4*(-1)**n/(2*n+1)
    #calcul des n termes suivant et comparaison
    stable = True
    print("\nTest pour n=",n)
    for k in range(n+1,2*n+1):
        Sni = 0
        for kk in range(k+1):
            Sni = Sni + 4*(-1)**kk/(2*kk+1)
        print(Sn,Sni)  
        if round(Sn,2) != round(Sni,2):
            stable = False
            break #pour casser la boucle en k
print(n)
```

# Exercice 8 - Suite de Fibonacci

Leonardo Fibonacci, un mathématicien italien du XIII\ieme\ siècle, a proposé le problème récréatif suivant dans l’un de ses ouvrages.

Un homme met un couple de lapins dans un lieu isolé de tous les côtés par un mur. Combien de couples obtient-on en un an si chaque couple engendre tous les mois un nouveau couple à compter du troisième mois de son existence ?

On en déduit les règles suivantes :
* au mois zéro, il n’y a aucun lapin ;
* au début du premier mois, il y a juste un couple de lapereaux ;
* les lapereaux ne procréent qu'à partir du début du troisième mois ;
* chaque début de mois, tout couple susceptible de procréer engendre effectivement un nouveau couple  de lapereaux ;
* les lapins ne meurent jamais.


Posons $n$ le numéro du mois et $F_n$ le nombre de couples de lapins au mois $n$ :

* $i  = 0$ ; $F_0 = 0$ ; 0 couple de lapins ; 
* $i  = 1$ ; $F_1 = 1$ ; 1 couple de lapins (petits) ; 
* $i  = 2$ ; $F_2 = 1$ ; 1 couple de lapins (moyens)  ; 
* $i  = 3$ ; $F_3 = 2$ ; 2 couples de lapins (1 grand, 1 petit) ; 
* $i  = 4$ ; $F_4 = 3$ ; 3 couples de lapins (1 grand, 1 moyen, 1 petit) ; 
* $i  = 5$ ; $F_5 = 5$ ; 5 couples de lapins (2 grands, 1 moyen, 2 petits) ; 
* $i = n$  ;  $F_n$ couples de lapins                                

!!! question Question 
A partir des itérations de 0 à 5 précédentes, déterminer une relation (dite de récurrence) entre $F_n$, $F_{n-1}$ et $F_{n-2}$.
!!!


!!! success Solution
On a simplement $F_n = F_{n-1}+F_{n-2}$ et $F_0=0$ et $F_1=1$.
!!!

!!! question Question
Écrire un programme qui à partir d'un nombre $n$ connu calcule et affiche le nombre de couples de lapins à chaque début de mois, du mois 0 au mois $n$ en utilisant la relation précédente.
!!!


```python
n = 5
Fn_2 = 0 #Fn-2 initialisé à 0 pour F0
Fn_1 = 1 #Fn-1 initialisé à 1 pour F1
if n <= 0 : print(0,Fn-2)
if n <= 1 : print(1,Fn_1)
for i in range(2,n+1):
    Fn = Fn_1 + Fn_2 #relation de récurrence
    Fn_2 = Fn_1      #Fn-2 prend la valeur Fn-1 pour l'itération suivante
    Fn_1 = Fn        #Fn-1 prend la valeur Fn
    print(i,Fn)
```

!!! question Question 
Tester le programme pour plusieurs valeurs de $n$ et indiquer au bout de combien de mois aura-t-on plus de 4000 couples de lapins, plus de 40000 couples de lapins ! 
!!!

!!! success Solution
Il y aura plus de 40000 lapins au 24ème mois.

# Exercice 9 - Remboursement d'emprunt

Lors d'un achat immobilier, il est important de déterminer la façon dont pourra être remboursé un crédit immobilier. En fonction de la somme empruntée, sur combien d'années s'étalera le prêt, quelles seront les mensualités de remboursement du crédit, quel taux pourra-t-on obtenir ?
Il existe de nombreux simulateurs en lignes permettant de faire ces différents calculs. Nous allons ici voir quels sont les codes en place derrière ces simulateurs.

On note $S$ la somme empruntée et $t_a$ le taux annuel (le taux mensuel est calculé simplement par la formule : $t_m=t_a/12$). On note $n$ le nombre de mois pour rembourser la somme empruntée et $m$ la mensualité constante.
Chaque mois, on rembourse une partie du capital emprunté ainsi que des intérêts proportionnel \textbf{au capital restant dû}. C'est là toute la difficulté de ce type d'emprunt. En notant $I_i$ le montant de l'intérêt chaque mois, le capital restant dû le mois $i$ est : $C_i=C_{i-1}-(m-I_i)$ avec $C_0=S$ et $I_i=C_{i-1}.t_m$.

!!! question Question
Ecrire un programme qui affiche le capital restant dû chaque mois et les intérêts du mois pour une somme $S$ fixée, un taux annuel fixé $t_a$ et une mensualité constante $m$. On testera le programme avec $S=400000$, $t_a={0.02}$ et $m={2000}$.
!!!


```python
S = 400000
ta = 0.02
tm = ta / 12
m = 2000

mois = 1
while S >=0:
    I = tm*S
    S = S - (m - I)
    print("mois",mois,"\t Capital restant:",round(S,2), "\tIntérêt du mois:",round(I,2))
    mois = mois + 1
```

!!! question Question
Ajouter au programme l'affichage à la fin du nombre de mois nécessaires pour rembourser le prêt (somme empruntée avec intérêts) ainsi que le total des intérêts versés. Améliorer le programme en arrondissant les intérêts au centime d'euro et en gérant la dernière mensualité différemment des autres.
!!!

```python
S = 400000
ta = 0.02
tm = ta / 12
m = 2000

mois = 1
It = 0
while S >0:
    I = round(tm*S,2)
    S = S - (m - I)
    if S<=0 : S = 0 #dernière mensualité plus petite en fait
    print("mois",mois,"\t Capital restant:",round(S,2), "\tIntérêt du mois:",round(I,2))
    mois = mois + 1
    It = It + I

print("Nombre de mois",mois-1,"Intérêts payés",It)
```

Dans le cas où le nombre de mois $n$ est fixé, on peut calculer la mensualité par la formule : $$m=\frac{{(1+t_m)}^n}{({1+t_m)}^n-1}.S.t_m$$
Toutefois la mensualité est arrondie au centime d'euro supérieur. La dernière mensualité est donc ainsi réduite.
!!! question Question 
Dans le cas où le nombre de mois $n$ est fixé (la somme empruntée $S$ est également fixée), réaliser un programme qui calcule la mensualité et l'arrondit au centime d'euro supérieur. Calculer la dernière mensualité et indiquer le total des intérêts.
!!!

```python
S = 400000
ta = 0.02
tm = ta / 12

mois = 1
It = 0
while S >0:
    I = round(tm*S,2)
    S = S - (m - I)
    if S<=0 : 
        m_dernier = m+S #dernière mensualité plus petite en fait
        S = 0 
    print("mois",mois,"\t Capital restant:",round(S,2), "\tIntérêt du mois:",round(I,2))
    mois = mois + 1
    It = It + I

print("Nombre de mois",mois-1,"Intérêts payés",It,"Denière mensualité",m_dernier)
```
