!!! question Question 3
Construire une fonction qui prend en entrée un entier naturel $n$ non nul et renvoie $\displaystyle s=\sum_{k=1}^{n} \frac{1}{k}$. Si cette fonction est appliquée à un entier négatif ou nul, elle affichera un message d'erreur et ne renverra rien. Tester avec $n=5$, on doit obtenir $2.2833$.
!!!

````python
def somme(n):
    if n <=0 :
        print("n doit être positif strictement")
        return
    s = 0
    for k in range(1,n+1):
        s = s + 1/k
    return s
print(somme(5))
````

2 x + 1 & \textrm{ si } x > 3
\end{array} \right. $. 
Définir la fonction $f$ puis une nouvelle fonction $P(n)$ qui prend pour argument un entier $n\in\mathbb{N}$ et renvoie la valeur de $\displaystyle P=\prod_{k=-n}^{n}f(k)$. Tester pour $n=5$, on doit obtenir $1188000$.

\'Ecrire une fonction ``racine2ndDegre(a,b,c)`` qui renvoie les racines de l'équation $ax^2+b x+c=0$ en fonction des valeurs de $a \neq 0$, $b$, $c$ réels. On supposera que $a$ est forcément différent de 0. On rappelle que pour écrire le nombre complexe $2+3i$ on tape en python $2+3*1j$. Tester la fonction pour $a=1, b=3, c=2$ qui doit renvoyer $(-1.0, -2.0)$ puis $a=1, b=2, c=2$ qui doit renvoyer $((-1+1j), (-1-1j))$.
!!!

````python
def racine2ndDegre(a,b,c):
    import math
    assert a!=0, print("a doit être non nul")
    delta=b**2-4*a*c
    if delta>0:
        return (-b+math.sqrt(delta))/(2*a),(-b-math.sqrt(delta))/(2*a)
    elif delta==0:
        return -b/(2*a)
    else:
        return (-b+1j*math.sqrt(-delta))/(2*a),(-b-1j*math.sqrt(-delta))/(2*a)

print(racine2ndDegre(1,3,2))
print(racine2ndDegre(1,2,2))
````

!!! question Question 11
\'Ecrire une fonction ``combinaisons_des(n)`` qui renvoie le nombre de combinaisons permettant d'obtenir le nombre $n$ avec 3 dés de couleurs différentes dont les faces sont numérotées de 1 à 6. Par exemple pour $n=9$, il y a 29 combinaisons possibles.
!!!

````python
def combinaisons_des(n):
    #nb de combinaisons pour obtenir n avec 3 dés à 6 faces de 3 couleurs différentes
    combinaisons=0
    for i in range(1,7):
        for j in range(1,7):
            for k in range(1,7):
                som=i+j+k
                if som==n:
                    combinaisons+=1
    return combinaisons
    
print(combinaisons_des(9))
````

````python
def moyenne(L):
    S = 0
    for l in L:
        S = S + l
    return S/len(L)
print(moyenne(L), sum(L)/len(L))
````

Ecrire une fonction ``sommesCumulees(L)`` prenant en entrée une liste L et renvoyant la liste des sommes cumulées (depuis le premier terme). Par exemple, l'appel ``L=[0,1,2,3,4,5,6,7]`` puis ``sommesCumulees(L)`` doit renvoyer ``[0,1,3,6,10,15,21,28]``.

Evaluer le nombre d'additions réalisées en fonction de la longueur $n$ de la liste. Si ce nombre est de l'ordre de $n^2$, réécrire le programme pour arriver à un nombre de l'ordre de $n$.
!!!

````python
def sommeCumulees(L):
    Ls = [L[0]]
    for i in range(1,len(L)):
        Ls.append(Ls[-1]+L[i])
    return Ls
print(sommeCumulees([0,1,2,3,4,5,6,7]))
````
